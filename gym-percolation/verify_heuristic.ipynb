```python
import json
import matplotlib.pyplot as plt
```

```python
def plot_data(file_path):
    X1_train = []
    X2_train = []
    Y_train = []
    
    x_list = []
    
    print("Loading data...")
    try:
        with open(file_path) as json_file:
            json_data = json.load(json_file)
        for key, value in json_data.items():
            print("############ New Experiment #############")
            for index in range(len(value)):
                fig = plt.figure(figsize=(5,5))
                plt.imshow(value[index]['board_state'], origin='lower')
                for a in value[index]['action_space']:
                    plt.plot(a['y'], a['x'], 'bx')
                plt.plot(value[index]['action']['y'], value[index]['action']['x'], 'ro')
                plt.show()
            #break
        
        print("Done")
    except Exception as e:
        print(e)
        
file_path = './experiments/moveData/experiment_0.3_junk.json'
plot_data(file_path)
```
